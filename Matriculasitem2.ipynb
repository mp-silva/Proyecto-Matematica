```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('10_MATRICULAS_ED_SUPERIOR_NUBLE_2021.xlsx')
# Frecuencia de Género
freq_genero = df["GENERO"].value_counts()
print("Frecuencia por Género:\n", freq_genero, "\n")

# Frecuencia de Tipo de Institución
freq_institucion = df["TIPO DE INSTITUCION"].value_counts()
print("Frecuencia por Tipo de Institución:\n", freq_institucion, "\n")

# Frecuencia de Áreas de Conocimiento
freq_area = df["AREA CONOCIMIENTO"].value_counts()
print("Frecuencia por Área de Conocimiento:\n", freq_area, "\n")
```

```python
# Calcular promedio de arancel por área
prom_arancel = df.groupby('AREA CONOCIMIENTO')['VALOR ARANCEL (PESOS)'].mean()

# Colores variados para cada barra
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F', '#BB8FCE', '#85C1E9']

# Crear gráfico de barras con colores
prom_arancel.plot(kind='bar', figsize=(12, 6), color=colores, edgecolor='black')
plt.title('Promedio de Valor Arancel por Área de Conocimiento', fontsize=14)
plt.xlabel('Área de Conocimiento')
plt.ylabel('Valor Arancel (Pesos)')

# Configurar el eje Y con incrementos de 250,000 pesos
max_valor = int(prom_arancel.max()) + 250000
yticks_valores = range(0, max_valor, 250000)
plt.yticks(yticks_valores)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Mostrar valores
print("Promedio de arancel por área:")
print(prom_arancel.round(0))
```

```python
# Calcular promedio de matrícula por área
prom_matricula = df.groupby('AREA CONOCIMIENTO')['VALOR MATRICULA (PESOS)'].mean()

# Crear gráfico de barras
prom_matricula.plot(kind='bar', figsize=(12, 6), color='lightcoral', edgecolor='darkred')
plt.title('Promedio de Valor Matrícula por Área de Conocimiento', fontsize=14)
plt.xlabel('Área de Conocimiento')
plt.ylabel('Valor Matrícula (Pesos)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Mostrar valores
print("Promedio de matrícula por área:")
print(prom_matricula.round(0))
```

```python
# Calcular promedio de edad por tipo de institución
edad_prom = df.groupby('TIPO DE INSTITUCION')['EDAD'].mean()

# Colores variados para cada barra
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

# Crear histograma con colores
plt.figure(figsize=(10, 6))
edad_prom.plot(kind='bar', color=colores, edgecolor='black')
plt.title('Promedio de Edad por Tipo de Institución', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Institución')
plt.ylabel('Edad Promedio')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', alpha=0.3)

# Mostrar valores en las barras
for i, valor in enumerate(edad_prom):
    plt.text(i, valor + 0.1, f'{valor:.1f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Mostrar los valores
print("Promedio de edad por tipo de institución:")
print(edad_prom.round(1))
```

```python
# Calcular promedio de arancel por carrera
aranceles_carrera = df.groupby('NOMBRE CARRERA')['VALOR ARANCEL (PESOS)'].mean()
top_5_carreras = aranceles_carrera.nlargest(5)

# Colores variados para cada barra
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']

# Gráfico con colores
plt.figure(figsize=(10, 6))
top_5_carreras.plot(kind='barh', color=colores)
plt.title('Top 5 Carreras Más Caras')
plt.xlabel('Valor Arancel (Pesos)')

# Configurar el eje X con incrementos de 500,000 pesos
max_valor = int(top_5_carreras.max()) + 500000
plt.xticks(range(0, max_valor, 500000))

plt.tight_layout()
plt.show()

# Mostrar las carreras
print("Top 5 carreras más caras:")
for carrera, valor in top_5_carreras.items():
    print(f"• {carrera}: ${valor:,.0f}")
```
